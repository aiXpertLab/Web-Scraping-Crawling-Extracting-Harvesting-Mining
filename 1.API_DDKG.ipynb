```python
from fastbook import search_images_ddg
from fastdownload import download_url
from fastcore.all import L
from fastai.vision.all import download_images, Image, resize_images
from pathlib import Path
from time import sleep
```

```python
# scrape 1 bird
urls=search_images_ddg('bird photos', max_images=1)
download_url(urls[0], './img/bird.jpg')

im = Image.open('./img/bird.jpg')
im.to_thumb(256,256)
```

```python
# scrape 1 forest
img_forest='./img/forest.jpg'
download_url(search_images_ddg('forest photos', max_images=1)[0], img_forest)
Image.open(img_forest).to_thumb(256,256)
```

```python
# grab 8 examples of each of "bird" and "forest" photos, and save each group of photos to a different folder:
searches = 'forest','bird'
path = Path('./img/birds')

if not path.exists():
    for o in searches:
        dest = (path/o)
        dest.mkdir(exist_ok=True, parents=True)
        
        download_images(dest, urls= search_images_ddg(f'{o} photo')[:8])
        resize_images(dest, max_size=400, dest=dest)
```
